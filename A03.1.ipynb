# Paso 2, se usa el unicoendata para quitar acentos, se convierten -1, 1 mapeo 2

```python
import pandas as pd
import unicodedata

def norm(x):
    x = str(x).strip().lower()
    return ''.join(c for c in unicodedata.normalize("NFKD", x) if not unicodedata.combining(c))

Aguas = df.copy()

Aguas["TamanoN"] = Aguas["tamaño"].apply(norm)
Aguas["HieloN"]  = Aguas["hielo"].apply(norm)
Aguas["SaborN"]  = Aguas["sabor"].apply(norm)

Aguas["Tamano_bin_1"] = Aguas["TamanoN"].map({
    "grande": -1,
    "chica": 1
})

Aguas["Hielo_bin_1"] = Aguas["HieloN"].map({
    "con": 1,
    "sin": -1
})

Aguas["Agua_kahlua_bin_1"] = Aguas["SaborN"].map({
    "kahlua": 1,
    "kalhua": 1,      # por si viene así escrito
    "horchata": 0
}).fillna(0).astype(int)

Aguas["Agua_horchata_bin_1"] = Aguas["SaborN"].map({
    "kahlua": 0,
    "kalhua": 0,
    "horchata": 1
}).fillna(0).astype(int)

Aguas["Score"] = Aguas["score"]

tabla_binaria = Aguas[["Tamano_bin_1","Hielo_bin_1","Agua_kahlua_bin_1","Agua_horchata_bin_1","Score"]]
tabla_binaria
```

```python
import statsmodels.api as sm

X_bin = tabla_binaria[["Tamano_bin_1", "Hielo_bin_1", "Agua_kahlua_bin_1", "Agua_horchata_bin_1"]]
X_bin = sm.add_constant(X_bin)
y_bin = tabla_binaria["Score"]

resultados_binarios = sm.OLS(y_bin, X_bin).fit()

betas_binarios = resultados_binarios.params
print("Betas - codificación binaria (-1,1):")
print(betas_binarios)

comparacion_betas = pd.DataFrame({
    "Variable": betas_binarios.index,
    "Beta_binaria_menos1_1": betas_binarios.values,
})

print(comparacion_betas)
```

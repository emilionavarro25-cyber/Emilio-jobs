```python
import pandas as pd
import numpy as np

df = pd.read_excel("Boston Housing Dataset 1978.xlsx")
```

```python
df.describe()
```

```python
df.head()
```

```python
CRIMescalable = (df['CRIM'] - df['CRIM'].min()) / (df['CRIM'].max() - df['CRIM'].min())
CRIMescalable.describe()
```

```python
Scales = pd.DataFrame()
Scales["Var"] = np.array(["ZN", "INDUS", "CHAS", "NOX", "RM", "AGE", "DIS", "RAD", "TAX", "PTRATIO", "B", "LSTAT"])

Datasetescalable = pd.DataFrame()
for column in Scales["Var"]:
    Datasetescalable[column + "_escalable"] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())  
Datasetescalable.describe()
```

```python
Datasetescalable = pd.DataFrame()
for column in Scales["Var"]:
    Datasetescalable[column + "_escalable"] = (df[column] - df[column].min()) / (df[column].max() - df[column].min())  
Datasetescalable.describe()
```

```python
from sklearn.preprocessing import StandardScaler

import matplotlib.pyplot as plt

scaler_std = StandardScaler()
df['MEDV_std'] = scaler_std.fit_transform(df[['MEDV']])

plt.hist(df['MEDV_std'].dropna(), bins=30, edgecolor='k')
plt.title('Distribucin de MEDV estandarizada')
plt.xlabel('MEDV_std')
plt.ylabel('Frecuencia')
plt.show()
```
